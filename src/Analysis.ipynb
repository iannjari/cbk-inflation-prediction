```python
import os
import pandas as pd
```

```python
inflation=pd.read_csv('Inflation Rates.csv')
mobile=pd.read_csv('Mobile Payments.csv')
atms=pd.read_csv('ATMs.csv')
interbank_rates=pd.read_csv('Interbank Rates  Volumes.csv')
trade=pd.read_csv('Foreign Trade Summary.csv')
```

```python
inflation.dtypes
```

### Join the data

```python
look_up={'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,'October':10,
        'November':11,'December':12}

inflation['Month'] = inflation['Month'].apply(lambda x: look_up[x])

inflation['Month']=inflation['Year'].astype(str)+'-'+inflation['Month'].astype(str)

inflation=inflation.drop(columns='Year')

inflation.head(1)
```

```python
mobile['Month'] = mobile['Month'].apply(lambda x: look_up[x])
mobile['Month']=mobile['Year'].astype(str)+'-'+mobile['Month'].astype(str)
mobile=mobile.drop(columns='Year')
mobile.head(1)
```

```python
atms.columns
```

```python
atms['Month'] = atms['Month'].apply(lambda x: look_up[x])

atms['Month']=atms['Year'].astype(str)+'-'+atms['Month'].astype(str)

atms=atms.drop(columns='Year')

atms.head(1)
```

```python
print(interbank_rates['date_r'])
print('///')
print('Shape of df ',interbank_rates.shape)
```

```python
def get_month(date):
    return date.split('-')[1]

def get_year(date):
    return date.split('-')[0]

interbank_rates['Month'] = interbank_rates['date_r'].apply\
(lambda x:get_year(x)+'-'+get_month(x))

interbank_rates=interbank_rates.drop(columns=['date_r'])

interbank_rates.head(1)
```

```python
trade['Month']=trade['Year'].astype(str)+'-'+trade['Month'].astype(str)
trade.head(1)
```

```python

final_df=pd.merge(pd.merge(pd.merge(pd.merge(inflation,mobile,on='Month',how='outer'),interbank_rates,on='Month',how='outer'),atms,on='Month',how='outer'),trade,on='Month',how='outer')
final_df.head(1)
```
